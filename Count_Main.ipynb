```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
#import skimage
#import io
from skimage import color
from skimage import io
from skimage import filters
from skimage import measure
from skimage.morphology import disk
import skimage.morphology as sm


FG = cv2.imread("images/09-22-20201_s8c2x26033-652y26034-472m0.png")
n1 = len(FG)
n2 = len(FG[0])
FGRGB = cv2.cvtColor(FG, cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(FGRGB)
plt.title('Day0 Green', fontweight ="bold") 
plt.show() 
hsv = cv2.cvtColor(FG, cv2.COLOR_BGR2HSV)


labels_green=np.zeros([n1,n2])
low_green = np.array([0, 0, 200])
high_green = np.array([0, 0, 255])
mask_green = cv2.inRange(hsv, low_green, high_green)
locg = np.where(mask_green>0)
img_g = np.zeros([n1,n2,3]).astype(float)
img_g[locg[0],locg[1],1] = 255
labels_green[locg[0],locg[1]] = 255
# plt.figure()
# plt.imshow(img_g.astype(np.uint8))
# plt.title('Day0 Green P1', fontweight ="bold") 
# plt.show()


shuffle_green = color.rgb2gray(labels_green)
# plt.figure()
# plt.imshow(shuffle_green, cmap = plt.cm.gray)


shuffle_ = filters.median(shuffle_green, disk(1))
new_img_green = color.rgb2gray(shuffle_)
# plt.figure()
# plt.imshow(new_img, cmap = plt.cm.gray)

#new_img_green = color.rgb2gray(shuffle_green)

new_img_fat_green = sm.erosion(new_img_green,sm.square(3))

# plt.figure('filters',figsize=(20,20))


# plt.figure()
# plt.subplot(121)
# plt.title('fat image green')
# plt.imshow(new_img_fat_green, cmap = plt.cm.gray)

# plt.subplot(122)
# plt.title('fat green - orgin green')
# plt.imshow(new_img_fat_green - new_img_green)


contours_green = measure.find_contours(new_img_fat_green, 0.5)

fig, (ax0,ax1) = plt.subplots(1,2,figsize=(16,16))
ax0.imshow(new_img_fat_green,plt.cm.gray)
ax1.imshow(new_img_fat_green,plt.cm.gray)

for n, contour in enumerate(contours_green):
    ax1.plot(contour[:, 1], contour[:, 0], linewidth=2)
    print(contour[:,0])
    print(contour[:,1])
    print('ssssssss')
ax1.axis('image')
ax1.set_xticks([])
ax1.set_yticks([])
plt.show()
```
